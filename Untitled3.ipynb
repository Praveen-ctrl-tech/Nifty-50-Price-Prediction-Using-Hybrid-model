```python
# =============================================================
# NIFTY 50 HYBRID MODEL (ERROR-FREE VERSION)
# XGBoost + LSTM + Fusion Ensemble
# =============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# ML Models
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

# Deep Learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# yfinance for auto-updating data
import yfinance as yf


# =============================================================
# 1. AUTO-DOWNLOAD + AUTO-UPDATE NIFTY 50 DATA (25 YEARS)
# =============================================================

FILE_NAME = "nifty.csv"
TICKER = "^NSEI"   # Yahoo symbol for NIFTY 50 Index

def load_nifty_data():
    if os.path.exists(FILE_NAME):
        print("Loading saved dataset...")
        old_df = pd.read_csv(FILE_NAME)
        old_df["Date"] = pd.to_datetime(old_df["Date"])
        last_date = old_df["Date"].max().date()
        print("Last available date:", last_date)

        # Download only new rows
        new_df = yf.download(TICKER, start=last_date, interval="1d")
        if not new_df.empty:
            print("New data found. Updating dataset...")
            new_df.reset_index(inplace=True)
            updated = pd.concat([old_df, new_df], ignore_index=True)
            updated.drop_duplicates(subset="Date", keep="last", inplace=True)
            updated.to_csv(FILE_NAME, index=False)
            return updated

        print("No new data. Using existing file.")
        return old_df

    else:
        print("Downloading full 25-year dataset...")
        df = yf.download(TICKER, period="25y", interval="1d")
        df.reset_index(inplace=True)
        df.to_csv(FILE_NAME, index=False)
        return df


df = load_nifty_data()


# =============================================================
# 2. CLEAN + FEATURE ENGINEERING
# =============================================================
df = df[['Date','Open','High','Low','Close','Volume']].dropna()
df = df.sort_values("Date")
df.reset_index(drop=True, inplace=True)

df['Return'] = df['Close'].pct_change()
df['MA7'] = df['Close'].rolling(7).mean()
df['MA21'] = df['Close'].rolling(21).mean()

df = df.dropna().reset_index(drop=True)
print("Dataset shape:", df.shape)


# =============================================================
# 3. XGBOOST MODEL
# =============================================================
print("Training XGBoost...")

features = ['Open','High','Low','Volume','Return','MA7','MA21']
X = df[features]
y = df['Close']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8
)

xgb.fit(x_train, y_train)

df['pred_ml'] = xgb.predict(X)


# =============================================================
# 4. LSTM MODEL
# =============================================================
print("Training LSTM...")

seq_len = 60
prices = df['Close'].values

X_lstm = []
y_lstm = []

for i in range(seq_len, len(prices)):
    X_lstm.append(prices[i-seq_len:i])
    y_lstm.append(prices[i])

X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)
X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(seq_len,1)))
model.add(LSTM(32))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
model.fit(X_lstm, y_lstm, epochs=15, batch_size=32, verbose=1)

# Fill predictions
df['pred_lstm'] = np.nan
df.iloc[seq_len:, df.columns.get_loc('pred_lstm')] = model.predict(X_lstm).flatten()


# =============================================================
# 5. FUSION ENSEMBLE (ML + LSTM)
# =============================================================
print("Training Fusion Ensemble...")

fusion_features = df[['pred_ml','pred_lstm']].dropna().copy()
fusion_features.reset_index(drop=True, inplace=True)

fusion_target = df.loc[df['pred_lstm'].dropna().index, 'Close']
fusion_target.reset_index(drop=True, inplace=True)

fusion_model = LinearRegression()
fusion_model.fit(fusion_features, fusion_target)

df['final_pred'] = np.nan
df.loc[fusion_features.index, 'final_pred'] = fusion_model.predict(fusion_features)


# =============================================================
# 6. PERFORMANCE METRICS (FIXED VERSION)
# =============================================================
y_true = fusion_target
y_pred = df.loc[fusion_features.index, 'final_pred']

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))  # FIXED
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
r2 = r2_score(y_true, y_pred)

# Direction accuracy
direction_true = np.sign(np.diff(y_true))
direction_pred = np.sign(np.diff(y_pred))
directional_accuracy = np.mean(direction_true == direction_pred) * 100

print("\n================== MODEL PERFORMANCE ==================")
print("MAE  :", round(mae, 3))
print("RMSE :", round(rmse, 3))
print("MAPE :", round(mape, 3), "%")
print("R²   :", round(r2, 4))
print("Directional Accuracy :", round(directional_accuracy, 2), "%")
print("=======================================================\n")



# =============================================================
# 7. PLOT — ACTUAL VS HYBRID PREDICTION
# =============================================================
plt.figure(figsize=(12,6))
plt.plot(y_true.values, label="Actual", linewidth=2)
plt.plot(y_pred.values, label="Hybrid Prediction", linestyle="--")
plt.title("NIFTY 50 — Actual vs Hybrid Predicted Price")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()

print("Hybrid model completed successfully.")
```
